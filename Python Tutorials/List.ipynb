# **Python Lists**

Lists typically store **homogeneous data**, i.e., values of the same data type. They also may store **heterogeneous data**, i.e., data of many different types. However, Python consists of six data types which are capable to store the sequences, but the most common and reliable type is the list.

A list is a collection of values or items of different types. The items in the list are separated with the comma $(,)$ and enclosed with the square brackets **[ ]**.

### **Characteristics of Lists**

The list has the following characteristics:

*   The lists are ordered.
*   The element of the list can access by index.
*   The lists are the mutable type.
*   A list can store the number of various elements.

#### **List Index**

We can use the index operator **[ ]** to access an item in a list. In Python, indices start at 0. So, a list having five elements will have an index from 0 to 4. Trying to access indexes outside these will raise an $IndexError$. The index must only be an integer and not any other type like float or string. Else, it will result in $TypeError$.

![Indexing.png](https://cdn.programiz.com/sites/tutorial2program/files/python-list-index.png)

```python
# List Indexing.
c = [-45, 6, 0, 72, 1543]

print(c[1])  # Prints 6.
print(c[3])  # Prints 72.
print(c[4])  # Prints 1543.
print("\n")

# Indices Must Be Integers or Integer Expressions.
print(c[1 + 3])  # Prints the 4th element, i.e., c[4] = 1543.
print("\n")

# List elements may be used as variables in expressions.
print(c[0] + c[1] + c[2])  # Using List Elements in Expressions. Prints -39.
```

#### **Determine a List's length.**

```python
print(len(c))  # Prints 5.
```

#### **Nested Indexing**

```python
nested_list = ["Happy", [2, 0, 1, 5], 6, 7.5]

print(nested_list[0])     # Prints "Happy".
print(nested_list[1][3])  # Prints 5.
print(nested_list[2])     # Prints 6.
print(nested_list[3])     # Prints 7.5.
```

#### **Accessing Elements from the End of the List with Negative Indices**

Lists also can be accessed from the end by using ***negative indices***. Python allows negative indexing for its sequences. The index of -1 refers to the last item, -2 to the second last item, and so on.

#### **Indexing and Slicing List**

$list[start : stop : step]$

*   The $start$ denotes the starting index position of the list.
*   The $stop$ denotes the last index position of the list.
*   The $step$ is used to skip the $n^{th}$ element within a $start:stop$.

![image.png](https://i.ytimg.com/vi/KAXvMbD1Zac/maxresdefault.jpg)

```python
# List Indexing and Slicing in Python.

slice_list = ["A", "R", "I", "T", "R", "A", "G", "A", "N", "G", "U", "L", "Y"]

""" Includes elements at index 2, 3, and 4 but excludes elements at index 5. """
print("1.", slice_list[2:5])  # Prints ['I', 'T', 'R'].

""" Includes the last 5 elements. """
print("2.", slice_list[:-5])  # Prints ['N', 'G', 'U', 'L', 'Y'].

""" Includes element from 6th to end. """
print("3.", slice_list[5:])   # Prints ['A', 'G', 'A', 'N', 'G', 'U', 'L', 'Y'].

""" Includes all elements, i.e., beginning to end. """
print("4.", slice_list[:])

""" Slicing with Steps: The following code uses a step of 2 to create a slice with every other element of numbers. """
print("5.", slice_list[::2])  # We omitted the start and end indices, so 0 and len(numbers) are assumed, respectively.

""" Slicing with Negative Indices and Steps. """
print("6.", slice_list[::-1])

# This is equivalent to:
print("7.", slice_list[-1:-9:-1])
```

#### **Lists are Mutable.**

Lists are mutable, meaning their elements can be changed, unlike strings or tuples.

We can use the assignment operator $=$ to change an item or a range of items.

```python
# Correcting mistake values in a list.
odd = [2, 4, 6, 8]

# Change the 1st item.
odd[0] = 1
print(odd)

# Change 2nd to 4th items.
odd[1:4] = [3, 5, 7]
print(odd)
```

#### **Some Sequences are Immutable.**

Python's string and tuple sequences are immutable, i.e., they cannot be modified. We can get the individual characters in a string, but attempting to assign a new value to one of the characters causes a $TypeError$:

```python
s = ["Hello"]

print(s[0][0])  # Prints 'H'.

s[0][0] = "F"   # Returns TypeError.
```

#### **Adding Elements in the Lists**

We can add one item to a list using the $append()$ method or add several items using $extend()$ method.

```python
# Append and Extend lists in Python.
my_list = [25, 11, 1997, 6, 10]
my_list.append(1972)
print(my_list)

my_list.extend([11, 3, 1962])
print(my_list)
```

```python
sample_list = []
s = "abc"
sample_list.extend(s)
print(sample_list)

t = (1, 2, 3)
sample_list.extend(t)
print(sample_list)

"""
Rather than creating a temporary variable, like t, to store a tuple before appending it to a list, 
we might want to pass a tuple directly to extend. In this case, the tuple's parentheses are 
required, because extend expects one iterable argument:
"""

sample_list.extend((4, 5, 6))  # Note the extra parentheses.
print(sample_list)
```

We can also use the $+$ operator to combine two lists. This process is also called ***concatenation***. The $*$ operator repeats a list for the given number of times.

```python
# Concatenate Lists.
ami = [25, 11, 1997]
ma = [6, 10, 1972]
baba = [11, 3, 1962]

total = ami + ma + baba
print(total)

# Repeat Lists.
print(["aritra"] * 3)
```

**Append to a List with $+=$.**

Let's start with an empty list **[ ]**, then use a $for()$ loop and $+=$ to append the values 1 through 5 to the list. The list grows dynamically to accommodate each item:

```python
a_list = []
for number in range(1, 6):
    """ When the left operand of += is a list, the right operand must be iterable; otherwise, a TypeError occurs. """
    a_list += [number]
print(a_list)

# The following append the characters of "Python" to the list letters:
letters = []
letters += "Python"
print(letters)
# If the right operand of += is a tuple, its elements also are appended to the list.
```

**Append to a List with the $insert()$ method.**

The $insert()$ method inserts an element to the list at the specified index.

**Syntax:** The syntax of the $insert()$ method is $list.insert(i, item)$

Here, the "$item$" is inserted into the list at the $i^{th}$ index. All the elements after "$item$" are shifted to the right.

```python
num = [1, 4, 7, 10]
num.insert(2, 13)
print(num)
```

#### **Delete Elements in the Lists**

The $del$ statement removes elements from a list and deletes variables from the interactive session. We can remove the element at any valid index or the element(s) from any valid slice.

```python
# Delete List Items.
del_list = ["A", "R", "I", "T", "R", "A"]

# Delete One Element.
del del_list[2]
print(del_list)

# Delete Multiple Elements.
del del_list[1:3]
print(del_list)

# Delete Entire List.
del del_list
print(del_list)  # Error: List not defined.
```

We can also use the $remove()$ method to remove the given item or the $pop()$ method to remove an item at the given index.

The $pop()$ method removes and returns the last item if the index is not provided. We can also use the $clear()$ method to empty a list.

```python
# Delete List Items.
del_list = ["A", "R", "I", "T", "R", "A"]

del_list.remove("A")    # Removes only the first character 'A'.
print(del_list)         # ['R', 'I', 'T', 'R', 'A']

print(del_list.pop(1))  # Pops the second character (index 1) 'I'.
print(del_list)         # ['R', 'T', 'R', 'A']

print(del_list.pop())   # Pops the last element 'A'.
print(del_list)         # ['R', 'T', 'R']

del_list.clear()        # Clear the entire list.
print(del_list)
```

**Sorting Lists**

Sorting arranges data either in ascending or descending order.

```python
# Sort a List of Numbers in Ascending Order.
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
numbers.sort()  # List method sort() modifies a list to arrange its elements in ascending order.
print(numbers)

# Sort a String in Ascending Order.
letters = "fadgchjebi"
ascending_letters = sorted(letters)
print(ascending_letters)

# Sort a Tuple in Ascending Order.
colors = ("red", "orange", "yellow", "green", "blue")
ascending_colors = sorted(colors)
print(ascending_colors)
```

```python
# Sort a List of Numbers in Descending Order.
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
numbers.sort(reverse=True)
print(numbers)

# Sort a String in Descending Order.
letters = "fadgchjebi"
descending_letters = sorted(letters, reverse=True)
print(descending_letters)

# Sort a Tuple in Descending Order.
colors = ("red", "orange", "yellow", "green", "blue")
descending_colors = sorted(colors, reverse=True)
print(descending_colors)
```

### **Searching Sequences**

Searching is the process of locating a key.

List Method index takes as an argument a search key, i.e., the value to locate in the list, and then searches through the list from index 0 and returns the index of the first element that matches the search key.

```python
numbers = [3, 7, 1, 4, 2, 8, 5, 6]
numbers.index(5)
```

Specifying the starting and ending indices causes the index to search from the starting index up to but not including the ending index location.

### **List Comprehension**

List comprehension is an elegant and concise way to create a new list from an existing list in Python. A list comprehension consists of an expression followed by a $for()$ statement inside square brackets.

```python
# Construct output List WITHOUT Using List Comprehensions.

input_list = [1, 2, 3, 4, 4, 5, 6, 7, 7]

output_list = []

# Use a for() loop for constructing the output List.
for var in input_list:
    if var % 2 == 0:
        output_list.append(var)

print("Output List using for() loop:", output_list)

# Construct output List Using List Comprehensions.

list_using_comp = [var for var in input_list if var % 2 == 0]

print("Output List using List Comprehensions:", list_using_comp)
```

```python
# An example to make a list with each item with being increasing power of 2.

pow2 = [2**x for x in range(10)]
print(pow2)
```

```python
# This code is equivalent to:

pow2 = []
for x in range(10):
    pow2.append(2**x)
print(pow2)
```

**Filter: List Comprehensions with $if$ Clauses.**

A list comprehension can optionally contain more $for()$ or $if$ statements. An optional $if$ statement can filter out items for the new list.

```python
pow2 = [2**x for x in range(10) if x > 5]
print(pow2)

odd = [x for x in range(20) if x % 2 == 1]
print(odd)

[x + y for x in ["Python ", "C "] for y in ["Language", "Programming"]]
```

**List Comprehension That Processes Another List's Elements.**

```python
colors = ["Red", "orAnge", "yellow", "green", "BLUE"]

colors_upper = [item.upper() for item in colors]
print(colors_upper)

colors_lower = [item.lower() for item in colors]
print(colors_lower)
```

## **Two-Dimensional Lists**

Lists can contain other lists as elements. A typical use of such nested (or multidimensional) lists is to represent tables of values consisting of information arranged in rows and columns.

To identify a particular table element, we specify two indices $-$ by convention, i.e., the first indices identifies the element's row, and the second indices identify the element's column. Lists that require two indices to identify an element are called two-dimensional lists. Multi-dimensional lists can have more than two indices. Here, we introduce two-dimensional lists.

Screenshot (137).png

```python
"""
Creating a Two-Dimensional List.

Consider a two-dimensional list with three rows and four columns (i.e., a 3-by-4 list) 
that might represent the grades of three students who each took four exams in a course:
"""

a = [[77, 68, 86, 73], [96, 87, 89, 81], [70, 90, 86, 81]]

"""
Writing the list as follows makes its row and column tabular structure clearer:

a = [[77, 68, 86, 73],  # first student's grades.
     [96, 87, 89, 81],  # second student's grades
     [70, 90, 86, 81]]  # third student's grades

In the two-dimensional list 'a':
a) 77, 68, 86 and 73 initialize a[0][0], a[0][1], a[0][2] and a[0][3], respectively.
b) 96, 87, 89 and 81 initialize a[1][0], a[1][1], a[1][2] and a[1][3], respectively, and
c) 70, 90, 86 and 81 initialize a[2][0], a[2][1], a[2][2] and a[2][3], respectively.

A list with 'm' rows and 'n' columns is called an m-by-n list and has m × n elements.
"""

print(a)
```

```python
for row in a:
    for item in row:
        print(item, end=" ")
    print()
```

```python
# How the Nested Loops Execute?
# Let's modify the nested loop to display the list's name and the row and column indices and value of each element:
for i, row in enumerate(a):
    for j, item in enumerate(row):
        print(f"a[{i}][{j}]={item} ", end=" ")
    print()
```
